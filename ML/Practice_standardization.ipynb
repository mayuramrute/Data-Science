# StandardScaler 

```python
# standardization of data meance the mean should be at 0 and std wil be 1 
```

```python
from sklearn.preprocessing import StandardScaler   # importing scaler which scale the colvalues to standard form
```

```python
standard_scaler = StandardScaler()   # object creation of standard scaler to fit the data in specific range
```

```python
standard_scaler
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# fit the scaler to the train set, it will learn the parameters
standard_scaler.fit(X_train)

# transform train and test sets
X_train_scaled = standard_scaler.transform(X_train)
X_test_scaled = standard_scaler.transform(X_test)
```

```python
X_train_scaled # both train and test are scaled 
```

```python
X_test_scaled
```

```python
np.round(X_train.describe(),1)   # before scaling mean ans std
```

```python
np.round(X_train_scaled.describe(),1) # will not work need to conver first into DF
```

```python
X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train.columns)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=X_test.columns)
```

```python
X_train_scaled    # scaled data 
```

```python
fig, (ax1,ax2) = plt.subplots(ncols=2, figsize=(12,5))

ax1.scatter(X_train['Age'],X_train['EstimatedSalary'])
ax1.set_title('Before scaling')

ax2.scatter(X_train_scaled['Age'],X_train_scaled['EstimatedSalary'], color='green')
ax2.set_title('After scaling')

plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

# before scaling
ax1.set_title('Before Scaling')
sns.kdeplot(X_train['Age'], ax=ax1)
sns.kdeplot(X_train['EstimatedSalary'], ax=ax1)

# after scaling
ax2.set_title('After Standard Scaling')
sns.kdeplot(X_train_scaled['Age'], ax=ax2)
sns.kdeplot(X_train_scaled['EstimatedSalary'], ax=ax2)
plt.show()
```

```python
# comparision of Distribution
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

# before scaling
ax1.set_title('Age Distribution Before Scaling')
sns.kdeplot(X_train['Age'], ax=ax1)

# after scaling
ax2.set_title('Age Distribution After Standard Scaling')
sns.kdeplot(X_train_scaled['Age'], ax=ax2)
plt.show()
```

# Why scaling is important?

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
lr_scaled = LogisticRegression()
```

```python
lr.fit(X_train,y_train)
lr_scaled.fit(X_train_scaled,y_train)
```

```python
y_pred = lr.predict(X_test)
y_scaled_pred = lr_scaled.predict(X_test_scaled)
```

```python
y_pred.shape
```

```python
y_scaled_pred
```

```python
# checking the effect of accuracy before and after scaling 
```

```python
from sklearn.metrics import accuracy_score
# from sklearn.matrics import accuracy_score
```

```python
print('Actual', accuracy_score(y_test,y_pred))  # with test data and actual data
print('scaled', accuracy_score(y_test,y_scaled_pred))  # with test data data and scaled data
# print("Actual",accuracy_score(y_test,y_pred))
# print("Scaled",accuracy_score(y_test,y_pred_scaled))
```

```python
from sklearn.tree import DecisionTreeClassifier 
```

```python
dt = DecisionTreeClassifier()
dt_scaled =DecisionTreeClassifier()
```

```python
dt.fit(X_train,y_train) # two objects scaled and unscaled
dt_scaled.fit(X_train_scaled,y_train)
```

```python
y_pred_dt = dt.predict(X_test)
y_scaled_pred_dt = dt_scaled.predict(X_test_scaled)
```

```python
print('Actual with DT: ' , accuracy_score(y_test, y_pred_dt))
print('Scaled with DT: ', accuracy_score(y_test, y_scaled_pred_dt))
```

# Effect of Outlier

```python
# addinf few outlies in the df
```

```python
df = pd.concat([df, pd.DataFrame({'Age': [5, 90, 95], 'EstimatedSalary': [1000, 250000, 350000], 'Purchased': [0, 1, 1]})], ignore_index=True)
```

```python
df
```

```python
plt.scatter(df['Age'], df['EstimatedSalary'])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop('Purchased', axis=1),
                                                    df['Purchased'],
                                                    test_size=0.3,
                                                    random_state=0)

X_train.shape, X_test.shape
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# fit the scaler to the train set, it will learn the parameters
scaler.fit(X_train)

# transform train and test sets
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train.columns)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=X_test.columns)
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

ax1.scatter(X_train['Age'], X_train['EstimatedSalary'])
ax1.set_title("Before Scaling")
ax2.scatter(X_train_scaled['Age'], X_train_scaled['EstimatedSalary'],color='red')
ax2.set_title("After Scaling")
plt.show()
```
