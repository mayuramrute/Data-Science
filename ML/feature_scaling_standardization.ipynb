# mean normalization

```python
df2['Fare']
```

```python
mean_fare = df2['Fare'].mean()
min_fare = df2['Fare'].min()
max_fare = df2['Fare'].max()
```

```python
df2['Fare'] = (df2['Fare'] - mean_fare) / (max_fare - min_fare)
```
