```python
import numpy as np 
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('heart.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.iloc[:,:-1],df.iloc[:,-1], test_size=0.2, random_state=2)
```

```python
X_train.shape, X_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
```

```python
lor = LogisticRegression()
disT = DecisionTreeClassifier()
```

```python
lor.fit(X_train,y_train)
disT.fit(X_train,y_train)
```

```python
lor.fit(X_train,y_train)
```

```python
y_pred_lor=lor.predict(X_test)
y_pred_disT=disT.predict(X_test)
```

```python
y_pred_disT
```

```python
y_pred_disT
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
print('Accuarcy of logisticregrassion: ', accuracy_score(y_test, y_pred_lor))
print('Accuracy of DecisionTree: ', accuracy_score(y_test, y_pred_disT))
```

```python
confusion_matrix(y_test,y_pred_lor)
```

```python
confusion_matrix(y_test, y_pred_disT)
```

```python
pd.DataFrame(confusion_matrix(y_test,y_pred_lor), columns=list(range(0,2)))
```

```python
pd.DataFrame(confusion_matrix(y_test, y_pred_disT), columns=list(range(0,2)))
```

```python
result = pd.DataFrame()
```

```python
result['Actual Outpu']=y_test
result['Logistic Prediction']=y_pred_lor
result['Decision Tree']=y_pred_disT
```

```python
from sklearn.metrics import recall_score, precision_score, f1_score
```

```python
print('For Logistic Regression')
print('-'*50)
cmat = pd.DataFrame(confusion_matrix(y_test,y_pred_lor), columns=list(range(0,2)))
display(cmat)
print('-'*50)
print('Precision: ', precision_score(y_test,y_pred_lor))
print('Recall: ', recall_score(y_test,y_pred_lor))
print('F1 Score: ', f1_score(y_test,y_pred_lor))
```

```python
print('For DecisionTree')
print('-'*50)
cmat = pd.DataFrame(confusion_matrix(y_test,y_pred_disT), columns=list(range(0,2)))
display(cmat)
print('-'*50)
print('Precision: ', precision_score(y_test,y_pred_disT))
print('Recall: ', recall_score(y_test,y_pred_disT))
print('F1 Score: ', f1_score(y_test,y_pred_disT))
```
