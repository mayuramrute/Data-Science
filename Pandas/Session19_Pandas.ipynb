# agg() 
#### function is used to apply one or more aggregation operations to a DataFrame or Series. This method is especially useful when you want to perform multiple aggregate functions on one or more columns at once, often in combination with groupby()

```python
genres.agg('min')   # min of all groups column wise
```

```python
genres.agg({
    "Runtime":'mean',      # avg runtime 
    'IMDB_Rating':'mean',  # mean
    'No_of_Votes':'sum',   # sum of all votes groupwise
    'Gross': 'sum',        # gross sum
    "Metascore":'mean'     # mean
})

# here we apply diffirent aggrigation funtion on multipul columns at a single time 
```

```python
# passing List
genres.agg(['min','max','sum'])
```

```python
# specific operation on perticular columns
genres.agg({
    "Runtime":['min','max'],      # avg runtime 
    'IMDB_Rating':['min','mean'],  # mean
    'No_of_Votes':['min','max','sum'],   # sum of all votes groupwise
    'Gross': ['mean','sum'],        # gross sum
    "Metascore":'mean' 
})
```

## looping on groups
#### find the highest rating movies from each group

```python
# for group, data in genres:
#     display(data[data['IMDB_Rating']==data['IMDB_Rating'].max()])
df = pd.DataFrame(columns = movies.columns)
for group, data in genres:
    df.append(data[data['IMDB_Rating']==data['IMDB_Rating'].max()])
```

```python
df = pd.DataFrame(columns=movies.columns)
rows_to_concat = []   # empty list to take all looping DF
for group,data in genres:
    max_rated_movies = data[data['IMDB_Rating'] == data['IMDB_Rating'].max()]
    rows_to_concat.append(max_rated_movies)

df= pd.concat(rows_to_concat, ignore_index=True)
display(df)
len(df)
```

# apply()
#### this funtion is specifically used to apply user defined logic on the perticular column
####  It can be a lambda function, a user-defined function, or any callable.
#### function can be used to perform more complex operations, such as conditional operations, string manipulation, or applying machine learning models

```python
genres.apply(max)
```

```python
# applying used define function
def starts_withA(groups):
    #display(groups)
    #display(group['Series_Title'])
    print(groups['Series_Title'].str.startswith('A').sum())   #this gives the bool series having true if movie name starts with 'A'
    return groups[groups['Series_Title'].str.startswith('A')]
```

```python
genres.apply(starts_withA)
```

## find ranking of each movie in the group according to IMDB score

```python
genres['IMDB_Rating'].rank().sort_values(ascending=False)
```

```python
# def rank_movies(group):
#     group['Genre_rank']=group['IMDB_Rating'].rank(ascending=False)
#     return group
# making the changes in the actual data fram has been depricated in higher version of python 2.0+

def rank_movies(group):
    group_copy = group.copy()
    group_copy['Genre_rank']=group_copy['IMDB_Rating'].rank(ascending=False)
    return group_copy
```

```python
#genres.apply(rank_movies)
genres.apply(rank_movies,include_groups=False)
```

### find normalized IMDB rating group wise

```python
def normalize(group):
    group['Norm_rating'] = (group["IMDB_Rating"] - group['IMDB_Rating'].min())/ (group["IMDB_Rating"].max() - group['IMDB_Rating'].min())
    return group
```

```python
genres.apply(normalize, include_groups=False)
```
