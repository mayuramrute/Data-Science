# 5 number summary box plot 

```python
# minimum, maximum, Q1,Q3,IQR(inter quartile range Q3-Q1) median
```

```python
import numpy as np
```

```python
lst_marks=[45,32,56,75,89,54,32,89,90,87,67,54,45,98,99,67,74]
```

```python
minimun, Q1, median, Q3, maximum = np.quantile(lst_marks,[0,0.25,0.50,0.75,1.0])
```

```python
minimun, maximum,Q1,Q3,median
```

```python
IQR = Q3-Q1
print(IQR)
```

```python
lower_fence = Q1-1.5*(IQR)
higher_fence = Q3+1.5*(IQR)
```

```python
lower_fence,higher_fence
```

```python
import seaborn as sns
```

```python
sns.boxenplot(lst_marks)
```
